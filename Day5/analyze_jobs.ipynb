```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# from google.colab import files
# files.upload()

data=pd.read_csv('jobs.csv')

data.head()
```

```python
#locations relative frequency histogram
filtered=data[data['Locations']!='Locations']
filtered['Locations'].value_counts(normalize=True).head(5).plot(kind='pie')
# as we can see literally 90% of the jobs posted are from Yeravan
```

```python
#Companies relative frequency histogram from a sample to make the visualization more appealing

data['Company Title'].value_counts().head(10).plot(kind='bar')
# as we can see from the plot below , SoftConstruct has the most amount of jobs posted followed by Digitain
```
